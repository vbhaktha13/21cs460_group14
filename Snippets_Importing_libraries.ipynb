````
!pip install streamlit
import streamlit as st
import sqlite3
import os
import re
from PyPDF2 import PdfReader

BASE_DIR = "CO2_Catalysis_Papers"
os.makedirs(BASE_DIR, exist_ok=True)

# Connect to SQLite database
conn = sqlite3.connect("co2_catalysis.db")
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE IF NOT EXISTS papers (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT,
    authors TEXT,
    year INTEGER,
    doi TEXT,
    keywords TEXT,
    filepath TEXT,
    abstract TEXT,
    faradaic_efficiency TEXT,
    catalyst_composition TEXT,
    electrochemical_conditions TEXT,
    products TEXT
)
""")
conn.commit()

def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages[:5]:  # limit to first 5 pages
        text += page.extract_text() + "\n"
    return text

def extract_abstract(text):
    match = re.search(r"Abstract(.*?)(Introduction|Methods|Results|Discussion)", text, re.S | re.I)
    if match:
        return match.group(1).strip()
    return "Abstract not found."

def extract_structured_info(text):
    data = {
        "Faradaic Efficiency (%)": [],
        "Catalyst Composition": [],
        "Electrochemical Conditions": [],
        "Products": []
    }

    # Faradaic Efficiency
    fe_matches = re.findall(r"(\d{1,3}\s?%)\s?(Faradaic efficiency|FE)", text, re.I)
    data["Faradaic Efficiency (%)"] = [m[0] for m in fe_matches]

    # Catalyst Composition
    catalyst_matches = re.findall(r"\b(Cu|Ag|Ni|Fe|Co|Zn|MOF|nanoparticle|alloy)\b", text, re.I)
    data["Catalyst Composition"] = list(set(catalyst_matches))

    # Electrochemical Conditions
    conditions = re.findall(r"(-?\d+\.\d+\s?V|\d+\s?mA/cm2|[0-9\.]+\s?M\s?[A-Za-z0-9]+)", text)
    data["Electrochemical Conditions"] = conditions

    # Products
    products = re.findall(r"\b(CO|CH4|C2H4|methanol|formate|ethanol)\b", text, re.I)
    data["Products"] = list(set(products))

    return data

# Sidebar navigation
st.sidebar.title("CO₂ Catalysis Paper Database")
menu = st.sidebar.radio("Navigation", ["Add Paper", "Search Papers", "View Database"])

# Add Paper
if menu == "Add Paper":
    st.header("Add a New Paper")
    title = st.text_input("Title")
    authors = st.text_input("Authors")
    year = st.number_input("Year", min_value=1900, max_value=2100, step=1)
    doi = st.text_input("DOI (optional)")
    keywords = st.text_input("Keywords (comma-separated)")
    pdf_file = st.file_uploader("Upload PDF", type=["pdf"])

    if st.button("Save Paper"):
        if pdf_file is not None:
            save_path = os.path.join(BASE_DIR, pdf_file.name)
            with open(save_path, "wb") as f:
                f.write(pdf_file.read())

            text = extract_text_from_pdf(save_path)
            abstract = extract_abstract(text)
            structured = extract_structured_info(text)

            cursor.execute("""
                INSERT INTO papers (title, authors, year, doi, keywords, filepath, abstract,
                                    faradaic_efficiency, catalyst_composition, electrochemical_conditions, products)
                VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
            """, (title, authors, year, doi, keywords, save_path, abstract,
                  "; ".join(structured["Faradaic Efficiency (%)"]),
                  "; ".join(structured["Catalyst Composition"]),
                  "; ".join(structured["Electrochemical Conditions"]),
                  "; ".join(structured["Products"])
            ))
            conn.commit()
            st.success("Paper added successfully with structured extraction!")

# Search Papers
elif menu == "Search Papers":
    st.header("Search Papers")
    keyword = st.text_input("Enter keyword (e.g., copper, electrocatalysis)")
    year = st.number_input("Filter by year (optional)", min_value=1900, max_value=2100, step=1, value=0)

    query = "SELECT * FROM papers WHERE keywords LIKE ?"
    params = [f"%{keyword}%"]
    if year > 0:
        query += " AND year=?"
        params.append(year)

    if st.button("Search"):
        results = cursor.execute(query, params).fetchall()
        for r in results:
            st.write(f"**Title:** {r[1]}")
            st.write(f"**Authors:** {r[2]}")
            st.write(f"**Year:** {r[3]}")
            st.write(f"**DOI:** {r[4]}")
            st.write(f"**Keywords:** {r[5]}")
            st.write(f"**Abstract:** {r[7]}")
            st.write(f"**Faradaic Efficiency:** {r[8]}")
            st.write(f"**Catalyst Composition:** {r[9]}")
            st.write(f"**Electrochemical Conditions:** {r[10]}")
            st.write(f"**Products:** {r[11]}")
            st.write(f"[Open PDF]({r[6]})")
            st.markdown("---")

# View Database
elif menu == "View Database":
    st.header("All Papers in Database")
    results = cursor.execute("SELECT * FROM papers").fetchall()
    for r in results:
        st.write(f"**Title:** {r[1]} | **Year:** {r[3]} | **DOI:** {r[4]}")
        st.write(f"**Faradaic Efficiency:** {r[8]}")
        st.write(f"**Catalyst Composition:** {r[9]}")
        st.write(f"**Electrochemical Conditions:** {r[10]}")
        st.write(f"**Products:** {r[11]}")
        st.markdown("---")
````
